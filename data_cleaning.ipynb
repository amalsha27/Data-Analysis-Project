```python
df['Date Local'] = pd.to_datetime(df['Date Local'], errors='coerce')

numeric_cols = ['CO', 'SO2', 'NO2', 'O3', 'PM25',
                'BarometricPressure', 'Temperature', 'WindResultant', 'DailyAQI']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Remove rows with NaN in the target or feature columns
df = df.dropna(subset=['CO', 'SO2', 'NO2', 'O3', 'PM25',
                             'BarometricPressure', 'Temperature', 'WindResultant', 'DailyAQI'])
```

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

```python
df = df[(df['Temperature'] > -50) & (df['Temperature'] < 60)]
```
